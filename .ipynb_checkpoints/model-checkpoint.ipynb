```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def get_win_pct(df):
    data = df[["ORtg", "DRtg"]]
    avg_diff = np.mean(df["ORtg"]) - np.mean(df["DRtg"])
    covar = np.cov(df["ORtg"], df["DRtg"])[0][1]
    denom = np.sqrt(np.var(df["ORtg"]) + np.var(df["DRtg"]) - 2*covar)
    value = avg_diff/denom
    final = round(norm.cdf(value), 3)
    return final
```

```python
def actual_win(df):
    win_count = 0
    lose_count = 0
    data = df[["Tm", "Opp.1"]]
    for i in range(len(data)):
        if data.iloc[i]["Tm"] > data.iloc[i]["Opp.1"]:
            win_count += 1
        else:
            lose_count += 1
    return round(win_count/(win_count+lose_count), 3)
```

```python
def rolling_exp(df):
    exp_win = np.zeros(len(df))
    
    for i in range(len(df)):
        exp_win[i] = get_win_pct(df.iloc[0:i+1])
    
    df["Rolling Win%"] = exp_win
    return exp_win
```

```python
teams = ["ATL", "BRK", "BOS", "CHI", "CHO", "CLE", "DAL", "DEN", "DET", "GSW", "HOU", "IND", "LAC", "LAL", "MEM",
           "MIA", "MIL", "MIN", "NOP", "NYK", "OKC", "ORL", "PHI", "PHO", "POR", "SAC", "SAS", "TOR", "UTA", "WAS"]
seasons = ["2017", "2018", "2019", "2020", "2021", "2022"]
ls = []

for team in teams:
    team_ls = []
    for season in seasons:
        df = pd.read_csv("game_logs/{}_{}.csv".format(team, season))
        rolling_exp(df)
        dfdict = {"Team": "{}_{}".format(team, season), "ORtg": round(np.mean(df["ORtg"]),2),
                  "DRtg": round(np.mean(df["DRtg"]), 2), "Expected Win%": get_win_pct(df), "Actual Win%": actual_win(df),
                  "Expected Change": "-"}

        if dfdict["Actual Win%"] > dfdict["Expected Win%"] + np.nanstd(df["Rolling Win%"]):
            dfdict["Expected Change"] = "Down"
        elif dfdict["Actual Win%"] < dfdict["Expected Win%"] - np.nanstd(df["Rolling Win%"]):
            dfdict["Expected Change"] = "Up"

        team_ls.append(dfdict)
        
    # Check results
    for i in range(len(team_ls)-1):
        if (team_ls[i]["Expected Change"] == "Up" and team_ls[i]["Actual Win%"] <= team_ls[i+1]["Actual Win%"]) or (team_ls[i]["Expected Change"] == "Down" and team_ls[i]["Actual Win%"] > team_ls[i+1]["Actual Win%"]):
            team_ls[i]["Result"] = "True"
        elif (team_ls[i]["Expected Change"] == "Up" and team_ls[i]["Actual Win%"] > team_ls[i+1]["Actual Win%"]) or (team_ls[i]["Expected Change"] == "Down" and team_ls[i]["Actual Win%"] < team_ls[i+1]["Actual Win%"]):
            team_ls[i]["Result"] = "False"
        else:
            team_ls[i]["Result"] = "-"
        
        ls.append(team_ls[i])
    
exp_change_df = pd.DataFrame(ls)
```

```python
ind = exp_change_df[exp_change_df["Expected Change"] == "Down"].index
ind1 = ind + 1
down = set(list(ind)+list(ind1))
tels = []

for i in down:
    tels.append(exp_change_df.iloc[i])

exp_down = pd.DataFrame(tels)
```

```python
plt.figure(figsize=(15, 15))
plt.axis("off")
plt.table(cellText=exp_down.values, 
          colLabels=exp_down.columns,
          loc='center')
plt.savefig('expected_down.png')
plt.show()
```

```python
ind = exp_change_df[exp_change_df["Expected Change"] == "Up"].index
ind1 = ind + 1
down = set(list(ind)+list(ind1))
tels = []

for i in down:
    tels.append(exp_change_df.iloc[i])

exp_up = pd.DataFrame(tels)
```

```python
plt.figure(figsize=(15, 15))
plt.axis("off")
plt.table(cellText=exp_up.values, 
          colLabels=exp_up.columns,
          loc='center')
plt.savefig('expected_up.png')
plt.show()
```
